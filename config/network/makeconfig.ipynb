```python
import json
```

```python
with open("settings.json", "r") as f:
    settings = json.load(f)

len(settings["dnns"])
```

```python
for name, dnn in settings["dnns"].items():
    layers = "\n".join([f"  - {{name: {layer}, num: 500}}" for layer in dnn["layers"]])
    config_text = f"""# @package _group_

name: {dnn['dir']}
layers:
{layers}
"""
    print(config_text)

    with open(name + ".yaml", "w") as f:
        f.write(config_text)
```
